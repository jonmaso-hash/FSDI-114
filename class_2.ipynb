## Stacks
A stack is an ordered collection of elements where items are added and removed from one end, commonly reffered to as the "top", the oredering prinicple is reffered to as LIFO (Last in, First out)

##Queues
A queues is an ordered collection of elements where items are added and removed from opposite ends, commonly reffred to as the "front" and "back" or "rear" of the queue. The ordering prinipile is reffered to as FIFO (first in, first out) "first come, first served"
