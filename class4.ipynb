```python
# challenge 1
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


# challenge 2
class LinkedList:
    def __init__(self):
        self.head = None

    # challenge 3: append method
    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node

    # challenge 3.1 print_list method
    def print_list(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return

        while current is not None:
            print(current.value)
            current = current.next

    # challenge 4: count nodes
    def count_nodes(self):
        count = 0
        current = self.head

        while current is not None:
            count += 1
            current = current.next

        return count

    # insert at the beginning
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    # challenge 5: remove method
    def remove(self, value):

        # if list is empty
        if self.head is None:
            return

        # if the node to remove is the head
        if self.head.value == value:
            self.head = self.head.next
            return

        # search for node before the one to remove
        current = self.head
        while current.next is not None:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next


# Create a linked list
sll = LinkedList()

# append values
sll.append(10)
sll.append(15)
sll.append(20)
sll.append(25)
sll.append(30)

print("Initial singly linked list:")
sll.print_list()

print("Node count:", sll.count_nodes())

# prepend
sll.prepend(50)

# remove value
sll.remove(25)

print("\nAfter prepend(50) and remove(25):")
sll.print_list()

print("Node count:", sll.count_nodes())
```

```python
# challenge 1: Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None


# challenge 2: Doubly Linked List class
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # challenge 3: append method
    def append(self, value):

        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return

        current = self.head

        while current.next is not None:
            current = current.next

        current.next = new_node
        new_node.prev = current

    # challenge 4: print_forward method
    def print_forward(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return

        while current is not None:
            print(current.value)
            current = current.next

    # challenge 5: print_backward method
    def print_backward(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return

        # move to tail
        while current.next is not None:
            current = current.next

        # walk backwards
        while current is not None:
            print(current.value)
            current = current.prev


# testing classes and methods
dll = DoublyLinkedList()

# add values
dll.append("A")
dll.append("B")
dll.append("C")
dll.append("D")

print("Print forward:")
dll.print_forward()

print("Print backward:")
dll.print_backward()
```
