```python
nums = [12, 3, 156, 1000, 55, 2]

def find_largest(nums):
    if not nums:
        return "no values to compare"
    
    largest = nums[0]

    for n in nums[1:]:
        if n > largest: 
            largest = n
    
    return largest 

print(find_largest(nums))
```
