```python
# challenge 1: Create the stack Class
class Stack:
    def __init__(self):
        self.items = []

    # check if the stack is empty
    def is_empty(self):
        return len(self.items) == 0

    # push() method
    def push(self, item):
        self.items.append(item)

    # pop() method
    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()

    # peek() method
    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

    # size() method
    def size(self):
        return len(self.items)


#create a stack
stack1 = Stack()

#push three items 
stack1.push(5)
stack1.push(10)
stack1.push(15)

#Peek at the top item 
print("stack peek:", stack1.peek())

#pop the top item
print("Stack pop:", stack1.pop())

#check the size
print("stack size:", stack1.size())
```

```python
# challenge 1: Create the Queue Class
class Queue:
    def __init__(self):
        self.items = []

    # is_empty() method
    def is_empty(self):
        return len(self.items) == 0
    
    # enqueue() method
    def enqueue(self, item):
        self.items.append(item)

    # dequeue() method
    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.pop(0)
    
    # size() method
    def size(self):
        return len(self.items)


# create queue
queue1 = Queue()

queue1.enqueue(20)
queue1.enqueue(40)
queue1.enqueue(60)

print("Queue dequeue:", queue1.dequeue())
print("Queue size:", queue1.size())
print("Is empty?", queue1.is_empty())


```
